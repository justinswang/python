```python
# the regular way
def get_counts(sequence):
    counts={}
    for x in sequence:
        if x in counts:
            counts[x] += 1
        else:
            counts[x] = 1
    return counts
```

```python
counts = get_counts(time_zones)
counts['America/New_York']
```

```python
from collections import defaultdict
```

```python
def get_counts2(sequence):
    counts=defaultdic(int) # values will initialize to 0
    for x in sequence:
        counts[x] += 1
    return counts
```

```python
counts2 = get_counts(time_zones)
counts2['America/New_York']
```

```python
#top 10 time zones
def top_counts(count_dict, n=10):
    value_key_pairs =[(count,tz) for tz, count in count_dict.items()]
    value_key_pairs.sort()
    return value_key_pairs[-n:]
```

```python
top_counts(counts)
```

```python
# collections.Counter class
from collections import Counter
counts3 = Counter(time_zones)
counts3.most_common(10)
```

```python
cframe = frame[frame.a.notnull()]
o_s = np.where(cframe['a'].str.contains('Windows'),'Windows', 'Not Windows')
o_s[:10]
```

```python
by_tz_os =cframe.groupby(['tz',o_s])
agg_counts = by_tz_os.size().unstack().fillna(0)
agg_counts[:10]
```

```python
indexer = agg_counts.sum(1).argsort()
indexer[:10]
```

```python
count_subset =agg_counts.take(indexer)[-10:]
count_subset
```

```python
count_subset.plot(kind='barh', stacked=True)
```
